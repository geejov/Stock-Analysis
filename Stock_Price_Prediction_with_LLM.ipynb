```python
import yfinance as yf
import pandas as pd
import ta

# Download historical data for AAPL from 2020-01-01 to 2023-01-01
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

# Create lagged features
for col in ['Close', 'Open', 'High', 'Low', 'Volume']:
    for i in [1, 5, 10]:
        data[f'{col}_lag_{i}'] = data[col].shift(i)

# Calculate Moving Averages
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

data['EMA_5'] = data['Close'].ewm(span=5, adjust=False).mean()
data['EMA_10'] = data['Close'].ewm(span=10, adjust=False).mean()
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()

# Calculate Technical Indicators using 'ta' library, passing the Close column as a Series
data['RSI'] = ta.momentum.RSIIndicator(data['Close'].squeeze(), window=14).rsi()
data['MACD'] = ta.trend.MACD(data['Close'].squeeze()).macd()
data['MACD_signal'] = ta.trend.MACD(data['Close'].squeeze()).macd_signal()
data['MACD_hist'] = ta.trend.MACD(data['Close'].squeeze()).macd_diff()
data['Bollinger_High'] = ta.volatility.BollingerBands(data['Close'].squeeze()).bollinger_hband()
data['Bollinger_Low'] = ta.volatility.BollingerBands(data['Close'].squeeze()).bollinger_lband()
data['Bollinger_Mid'] = ta.volatility.BollingerBands(data['Close'].squeeze()).bollinger_mavg()

# Handle missing values
data.dropna(inplace=True)

# Display the first few rows with new features
print(data.head())
```

## Data preparation

### Subtask:
Prepare the data for the regression model by defining the target variable (tomorrow's stock price) and the features.


**Reasoning**:
Create the target variable by shifting the 'Close' price, drop the original 'Close' column and any resulting NaN values, and then define the feature and target sets.



```python
# Create the target variable (tomorrow's closing price)
data['target'] = data['Close'].shift(-1)

# Drop the original 'Close' column and rows with missing values (the last row will be NaN in 'target')
data.drop('Close', axis=1, inplace=True)
data.dropna(inplace=True)

# Define features (X) and target (y)
X = data.drop('target', axis=1)
y = data['target']

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
print("\nFirst few rows of features:")
display(X.head())
print("\nFirst few values of target:")
display(y.head())
```

## Data splitting

### Subtask:
Split the data into training and testing sets with a 70/30 ratio.


**Reasoning**:
Split the features and target into training and testing sets with a 70/30 ratio using train_test_split.



```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Model training

### Subtask:
Train a regression model (e.g., Linear Regression) on the training data.


**Reasoning**:
Train a Linear Regression model using the training data.



```python
from sklearn.linear_model import LinearRegression

# Instantiate the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the trained model on the testing data and report performance metrics like Mean Squared Error (MSE) and R-squared (RÂ²).


**Reasoning**:
Evaluate the trained Linear Regression model on the testing data by calculating the Mean Squared Error (MSE) and R-squared (RÂ²) score.



```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)

# Print the performance metrics
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (RÂ²): {r2}")
```

## Feature engineering

### Subtask:
Ensure the necessary features (Historical Prices, Volume, Moving Averages, Technical Indicators) are available and prepared.


**Reasoning**:
Reload the data and create the specified features including lagged prices, volume, moving averages, and technical indicators using the `ta` library, then handle missing values by dropping rows with NaNs.



```python
import yfinance as yf
import pandas as pd
import ta

# Download historical data for AAPL from 2020-01-01 to 2023-01-01
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")

# Create lagged features
for col in ['Close', 'Open', 'High', 'Low', 'Volume']:
    for i in [1, 5, 10]:
        data[f'{col}_lag_{i}'] = data[col].shift(i)

# Calculate Moving Averages
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

data['EMA_5'] = data['Close'].ewm(span=5, adjust=False).mean()
data['EMA_10'] = data['Close'].ewm(span=10, adjust=False).mean()
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()

# Calculate Technical Indicators using 'ta' library, passing the Close column as a Series
data['RSI'] = ta.momentum.RSIIndicator(data['Close'].squeeze(), window=14).rsi()
data['MACD'] = ta.trend.MACD(data['Close'].squeeze()).macd()
data['MACD_signal'] = ta.trend.MACD(data['Close'].squeeze()).macd_signal()
data['MACD_hist'] = ta.trend.MACD(data['Close'].squeeze()).macd_diff()
data['Bollinger_High'] = ta.volatility.BollingerBands(data['Close'].squeeze()).bollinger_hband()
data['Bollinger_Low'] = ta.volatility.BollingerBands(data['Close'].squeeze()).bollinger_lband()
data['Bollinger_Mid'] = ta.volatility.BollingerBands(data['Close'].squeeze()).bollinger_mavg()

# Handle missing values
data.dropna(inplace=True)

# Display the first few rows with new features
display(data.head())
```

## Data preparation

### Subtask:
Create a binary target variable based on whether the closing price increases or decreases on the next day.


**Reasoning**:
Create the binary target variable 'Price_Increase' by comparing the current day's closing price to the next day's closing price, then drop any rows with NaN values which would include the last row as it does not have a 'next day'.



```python
# Create the binary target variable 'Price_Increase'
data['Price_Increase'] = (data['Close'].shift(-1) > data['Close']).astype(int)

# Drop the last row which will have a NaN in 'Price_Increase'
data.dropna(inplace=True)

# Display the first few rows with the new target variable
display(data.head())
```

## Data splitting

### Subtask:
Split the data into training and testing sets using the same 70/30 ratio as before.


**Reasoning**:

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = data.drop('Price_Increase', axis=1)
y = data['Price_Increase']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Model training

### Subtask:
Train a Random Forest Classifier model on the training data.


**Reasoning**:
Train a Random Forest Classifier model using the training data.



```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate a RandomForestClassifier model
model = RandomForestClassifier(random_state=42)

# Fit the model to the training data
model.fit(X_train, y_train)
```

## Model evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Calculate and print performance metrics
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)

# Get feature importances
feature_importances = model.feature_importances_

# Create a pandas Series for feature importances
feature_importance_series = pd.Series(feature_importances, index=X_train.columns)

# Sort feature importances in descending order and print
print("\nFeature Importances (sorted):")
display(feature_importance_series.sort_values(ascending=False))
```
